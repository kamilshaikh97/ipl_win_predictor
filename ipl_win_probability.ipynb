```python
match_df = match.merge(total_score_df[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

```python
#some of the teams have changed the name, so we will replace names with recent names
#and some are not participating in recent events, so we will not consider them
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
#lets keep only those matches, which are currently playing 

match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

```python
# we can see that there are some matches which were rain affected and how would we find???
#where duckworth-lewis rule was applied!! big brain!

match_df.dl_applied.value_counts()
```

```python
#so there are 15 such matches, let's remove those matches
```

```python
match_df = match_df[match_df['dl_applied'] == 0]
```

```python
match_df.shape
```

```python
#we need match_id, city of the match, who won, and 1st inning total for our use from this dataframe 

match_df = match_df[['match_id','city','winner','total_runs']]
```

```python
delivery_df = match_df.merge(delivery,on='match_id')
```

```python
delivery_df.head()
```

```python
# okay, now we dont need to know how ball by ball 1st innings was build, so we we will only consider that info for 2nd inning 

delivery_df = delivery_df[delivery_df['inning'] == 2]
```

```python
delivery_df.head()
```

```python
# we need ball by ball, current score, we will do cumulative summation for that after doing groupby on match id
# we will also need how many runs need to be chased down, at that delivery 
#for that we will simply subtract current score from 1st inning total
```

```python
delivery_df['current_score'] = delivery_df.groupby('match_id').cumsum()['total_runs_y']
```

```python
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
delivery_df.head()
```

```python
delivery_df['balls_left'] = 126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df.head()
```

```python
#now we need one more very crucial information 
#that is how many wickets are left, after each ball, 
#we will extract this information from player dismissed column 
# we have NaN values when player did not get out and name of the player if player got out on that delivery
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
wickets_left = delivery_df.groupby('match_id').cumsum()['player_dismissed'].values
delivery_df['wickets_left'] = 10 - wickets_left
```

```python
delivery_df.tail()
```

```python
# cool 
# current run rate and required run rate usually plays great role in winning matches, so we will create those columns 
```

```python
# crr = runs/overs
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
delivery_df.crr
```

```python
delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
delivery_df.rrr
```

```python
# now we need a numeric column showing result, 1 if same team won which batted 1st and 0 if they lost 
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
#so that's pretty much it is, we created the columns we need for prediction 
#let's put only the required one in the final dataframe that we will be making model on 
```

```python
final_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets_left','total_runs_x','crr','rrr','result']]
```

```python
final_df.head()
```

```python
#but as we know, if we train data such that it is sequencial, it can create bias, so let's shuffle the data 
```

```python
final_df.shape[0]
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.head()
```

```python
final_df = final_df[final_df['balls_left'] != 0]
```

## data preprocessing for modelling 

```python
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
# we are going to use a lot of different algorithms so let's create a pipeline to ease and standardise our workflow
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
pipe.predict_proba(X_test)[10]
```
